DATA PREPERATION

```python
#drop columns CONTAINING NULL VALUES AND PROVIDE LITTLE USE
df_us_pollution.drop(['NO2 Units', 'SO2 Units','CO Units','County','Address','Site Num','NO2 1st Max Value'
                   ,'NO2 1st Max Hour','O3 Units','O3 1st Max Value','SO2 1st Max Value','SO2 1st Max Hour'
                   ,'CO 1st Max Value' ,'CO 1st Max Hour','O3 1st Max Hour','County Code','Unnamed: 0'], axis=1, inplace=True)
```

```python
df_us_deaths.drop(['measure', 'age','metric','upper','lower', 'cause', 'sex'], axis=1, inplace=True)
```

```python
df_us_pollution.head()
```

```python
#remove duplicate rows referring to same location and date containing NAN values
df_us_pollution.dropna(inplace=True)
```

```python
#convert string 'Date Local' to date
df_us_pollution["Date Local"] = pd.to_datetime(df_us_pollution["Date Local"])
```

```python
df_us_pollution['Year'] = df_us_pollution['Date Local'].dt.year
```

```python
df_us_pollution.head()
```

```python
#records for many states were poor in comparison with others
df_us_pollution['State'].value_counts()
```

```python
#in order to combine the datasets convert the data local to year only
df_us_pollution['Year'] = df_us_pollution['Date Local'].dt.year
df_us_pollution.drop(['City', 'Date Local'], axis=1, inplace=True)
```

```python
#drop columns that provide no benefit
df_us_pollution.drop(['NO2 AQI', 'SO2 AQI','CO AQI','O3 AQI','State Code'], axis=1, inplace=True)
```

```python
df_us_pollution.head()
```

MERGE THE DATASETS

```python
#GROUP ROWS TOGETHER ON STATE AND YEAR TO COMBINE POLLUTION DATASET WITH DEATHS
result_NO2  =df_us_pollution.groupby(['State', 'Year'])['NO2 Mean'].mean().reset_index()
result_SO2  =df_us_pollution.groupby(['State', 'Year'])['SO2 Mean'].mean().reset_index()
result_CO  =df_us_pollution.groupby(['State', 'Year'])['O3 Mean'].mean().reset_index()
result_O3  =df_us_pollution.groupby(['State', 'Year'])['CO Mean'].mean().reset_index()
df_mean_pollution_us = result_NO2
df_mean_pollution_us = pd.merge(df_mean_pollution_us, result_SO2, left_on = ['Year', 'State'], right_on = ['Year', 'State'])
df_mean_pollution_us = pd.merge(df_mean_pollution_us, result_CO, left_on = ['Year', 'State'], right_on = ['Year', 'State'])
df_mean_pollution_us = pd.merge(df_mean_pollution_us, result_O3, left_on = ['Year', 'State'], right_on = ['Year', 'State'])
```

```python
#sum up the number of deaths for each cause of death to a single row
df_us_deaths  =df_us_deaths.groupby(['location', 'year'])['val'].sum().reset_index()
```

```python
#rename columns to join datasets
df_us_deaths.rename(columns = {'location':'State', 'year':'Year'}, inplace = True)
```

```python
#merge datasets
df_pollutionANDdeaths_us = pd.merge(df_mean_pollution_us, df_us_deaths, left_on = ['Year', 'State'], right_on = ['Year', 'State'])
```

```python
df_pollutionANDdeaths_us
```

```python
df_pollutionANDdeaths_us.rename(columns = {'val':'Deaths'}, inplace = True)
```

```python
xTrain, xValid, yTrain, yValid=train_test_split(X,y,train_size=0.6, random_state=2)
```

```python
#normalize the data to scale the data to a number between 0 and 1 TO REDUCE THE VARIANCES IN THE DATA
#MAKING THE DATA MORE MANAGEABLE
from sklearn.preprocessing import MinMaxScaler
#apply to training set only
scaler=MinMaxScaler()
scaler.fit(xTrain)
```

```python
#assign created scaler values to the traing and test sets
xTrain=scaler.transform(xTrain)
xValid=scaler.transform(xValid)
```

```python
#AFTER NORMALIZING THE DATA
pd.DataFrame(xTrain).describe()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn import metrics
```

```python
#iterates 10 times updating the weights each time to improve the prediction through back propagation
nn=MLPRegressor(hidden_layer_sizes=(10,10,), activation='logistic', max_iter=10)
nn.fit(xTrain, yTrain)
```

```python
mae = metrics.mean_absolute_error(yTrain, nn.predict(xTrain))
mse = metrics.mean_squared_error(yTrain, nn.predict(xTrain))
rsq = metrics.r2_score(yTrain, nn.predict(xTrain))
print(mae, mse, rsq)
```

```python
#try different activation methods and number of hidden layers to try optimize the results further
nn=MLPRegressor(hidden_layer_sizes=(10,10,), activation='tanh', max_iter=1000)
nn.fit(xTrain, yTrain)
```

```python
mae = metrics.mean_absolute_error(yTrain, nn.predict(xTrain))
mse = metrics.mean_squared_error(yTrain, nn.predict(xTrain))
rsq = metrics.r2_score(yTrain, nn.predict(xTrain))
print(mae, mse, rsq)
```

```python
nn=MLPRegressor(hidden_layer_sizes=(10,10,10,), activation='relu', max_iter=1000)
nn.fit(xTrain, yTrain)
```

```python
mae = metrics.mean_absolute_error(yTrain, nn.predict(xTrain))
mse = metrics.mean_squared_error(yTrain, nn.predict(xTrain))
rsq = metrics.r2_score(yTrain, nn.predict(xTrain))
print(mae, mse, rsq)
```

```python
#check the validation data to make sure the network has not been overtrained
mae = metrics.mean_absolute_error(yValid, nn.predict(xValid))
mse = metrics.mean_squared_error(yValid, nn.predict(xValid))
rsq = metrics.r2_score(yValid, nn.predict(xValid))
print(mae, mse, rsq)
```

```python
col_names = ['State', 'Year', 'SO2 Mean', 'age', 'sex','val'] #label the columns
```

```python
regressionSet = pd.read_csv("newData.csv",header=None, names=col_names)
```

```python
regressionSet.head()
```

```python
regressionSet = regressionSet.iloc[1: , :]#remove NaN column
```

```python
regressionSet.head()
```

```python
regressionSet = regressionSet[regressionSet.sex != 'Both']#remove both gender rows
```

```python
labelencoder = LabelEncoder()#encode some values that are categorical and change columns to floats and int
```

```python
regressionSet['numericAge'] = labelencoder.fit_transform(regressionSet['age'])
```

```python
regressionSet['gender'] = labelencoder.fit_transform(regressionSet['sex'])
```

```python
regressionSet['SO2 Mean'] = regressionSet['SO2 Mean'].astype(float)
```

```python
regressionSet['val'] = regressionSet['val'].astype(float)
```

```python
regressionSet['gender'] = regressionSet['gender'].astype(int)
```

```python
regressionSet['numericAge'] = regressionSet['numericAge'].astype(int)
```

```python
regressionSet.dtypes
```

```python
regressionSet.head()
```

```python
newSet = regressionSet.drop(columns=['State','age','sex'])#make a new dataframe after droping extra columns
```

```python
setNowDeaths = newSet.drop(columns=['val'])#make a new dataframe without deaths
```

```python
setNowDeaths
```

```python
x = setNowDeaths.iloc[:,1:4].copy()#add age, gender and industrial pollution to x variable
```

```python
x
```

```python
y = newSet.iloc[:,2].copy()#add deaths to y variable
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)#split into train and test data
```

```python
#mse = try and lower variance in tree and best to make the best split of a node set the depth of the tree to a limit of 6 deep
regressor = DecisionTreeRegressor(criterion="mse",splitter="best", max_depth=6,random_state = 0).fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#I thoughtthe tree was overfitting the data the score was very hign. I was going to try and prune the tree but decided to look at the train and test accuracy
print('R Squared Score is:', r2_score(y_test, y_pred))
```

```python
#After looking at the test and train accuracy their isn't a large difference in the two scores so doesn't appear to be overfitting so decided not to prune the tree 
print('Training accuracy: ',regressor.score(X_train,y_train))
print('Test Accuracy: ',regressor.score(X_test,y_test))
```
